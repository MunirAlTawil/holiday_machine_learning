# DataTourisme POIs Exploration

This notebook explores the processed DataTourisme Point of Interest data.


```python
import pandas as pd
from pathlib import Path

# Load the processed CSV file
data_path = Path("../data/processed/datatourisme_pois.csv")
df = pd.read_csv(data_path)

print("Data loaded successfully!")
```

## Basic Information


```python
# Number of rows
print(f"Number of rows: {len(df)}")

# Columns
print(f"\nColumns ({len(df.columns)}):")
print(df.columns.tolist())
```

## Missing Values Summary


```python
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percent
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if len(missing_df) > 0:
    print("Missing values by column:")
    print(missing_df)
else:
    print("No missing values found!")
```

## Top Categories (if present)


```python
# Check if there's a category column
category_columns = [col for col in df.columns if 'categor' in col.lower() or 'type' in col.lower()]

if category_columns:
    for col in category_columns:
        print(f"\nTop 10 most frequent values in '{col}':")
        print(df[col].value_counts().head(10))
else:
    print("No category column found in the dataset.")
```
